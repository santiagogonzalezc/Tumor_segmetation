```python
import os
import random
from sklearn.model_selection import train_test_split
import shutil

main_folder = r'C:\Users\santi\Desktop\Tumor_segmetation\Data\nnUNet_raw\Dataset011_BrainTumour'
train_folder_path = os.path.join(main_folder, 'imagesTr')
test_folder_path = os.path.join(main_folder, 'imagesTs')

folders = [folder for folder in os.listdir(main_folder) if folder.startswith('UPENN') and os.path.isdir(os.path.join(main_folder, folder))]

train_folders, test_folders = train_test_split(folders, test_size=0.2, random_state=42)

for folder in train_folders:
    source_path = os.path.join(main_folder, folder)
    destination_path = os.path.join(train_folder_path, folder)
    shutil.move(source_path, destination_path)

for folder in test_folders:
    source_path = os.path.join(main_folder, folder)
    destination_path = os.path.join(test_folder_path, folder)
    shutil.move(source_path, destination_path)
```

```python
import os
import shutil

train_images_folder = r'C:\Users\santi\Desktop\Tumor_segmetation\Data\nnUNet_raw\Dataset011_BrainTumour\imagesTr'
masks_folder = r'C:\Users\santi\Desktop\Tumor_segmetation\Data\seg'
labels_folder = r'C:\Users\santi\Desktop\Tumor_segmetation\Data\nnUNet_raw\Dataset011_BrainTumour\labelsTr'

os.makedirs(labels_folder, exist_ok=True)

train_images = [os.path.splitext(image)[0] for image in os.listdir(train_images_folder)]


for mask in os.listdir(masks_folder):
    mask_name = os.path.splitext(mask)[0]

    
    for image_name in train_images:
        if mask_name.startswith(image_name):
            source_path = os.path.join(masks_folder, mask)
            destination_path = os.path.join(labels_folder, mask)
            shutil.move(source_path, destination_path)
            print(f"Movida la máscara: {mask} a {labels_folder}")
            break  

print(f"Todas las máscaras correspondientes a imágenes de entrenamiento se han movido a {labels_folder}")

```

```python
import os
import shutil

images_folder = r'C:\Users\santi\Desktop\Tumor_segmetation\Data\nnUNet_raw\Dataset011_BrainTumour\imagesTr'

modalities_map = {
    'FLAIR': '0000',
    'T1GD': '0002',
    'T1': '0001',
    'T2': '0003'
}

def get_subject_number(filename):
    try:
        subject_number = filename.split('-')[-1].split('_')[0]
        return subject_number
    except IndexError:
        return None

def get_modality(filename):
    for modality in modalities_map:
        if modality in filename.upper():
            return modality
    return None

for subject_folder in os.listdir(images_folder):
    subject_path = os.path.join(images_folder, subject_folder)
    
    if os.path.isdir(subject_path):
        
        for image_file in os.listdir(subject_path):
            image_path = os.path.join(subject_path, image_file)
            
            if os.path.isfile(image_path):
                
                subject_number = get_subject_number(image_file)
                
                if subject_number is None:
                    print(f"No se pudo extraer el número de sujeto de: {image_file}")
                    continue
                
                
                modality = get_modality(image_file)
                
                if modality is None:
                    print(f"No se pudo detectar la modalidad en el archivo: {image_file}")
                    continue
                
                
                new_filename = f"BrainTumour_{subject_number}_{modalities_map[modality]}.nii.gz"
                new_image_path = os.path.join(subject_path, new_filename)
                
                
                shutil.move(image_path, new_image_path)
                print(f"Renombrado: {image_file} a {new_filename}")

print("Renombrado completo.")
```

```python
import os
import shutil


images_folder = r'C:\Users\santi\Desktop\Tumor_segmetation\Data\nnUNet_raw\Dataset011_BrainTumour\imagesTr'

destination_folder = r'C:\Users\santi\Desktop\Tumor_segmetation\Data\nnUNet_raw\Dataset011_BrainTumour\imagesTr'
os.makedirs(destination_folder, exist_ok=True)

for subject_folder in os.listdir(images_folder):
    subject_path = os.path.join(images_folder, subject_folder)
    
    if os.path.isdir(subject_path):
        
        for image_file in os.listdir(subject_path):
            image_path = os.path.join(subject_path, image_file)
            
            if os.path.isfile(image_path):
                
                destination_path = os.path.join(destination_folder, image_file)
                shutil.move(image_path, destination_path)
                print(f"Movido: {image_file} a {destination_folder}")
       
        shutil.rmtree(subject_path)
        print(f"Carpeta del sujeto {subject_folder} eliminada.")

print("Proceso completo. Todas las imágenes se han movido y las carpetas de sujetos se han eliminado.")
```

```python
import nibabel as nib
import numpy as np
import os


input_dir = r'C:\Users\santi\Desktop\Tumor_segmetation\Data\nnUNet_raw\Dataset011_BrainTumour\labelsTr'

for filename in os.listdir(input_dir):
    if filename.endswith('.nii.gz'):
        file_path = os.path.join(input_dir, filename)


        mask = nib.load(file_path)


        mask_data = mask.get_fdata()


        mask_data[mask_data == 4] = 3


        new_mask = nib.Nifti1Image(mask_data, mask.affine, mask.header)


        nib.save(new_mask, file_path)

        print(f"Modificado: {filename}")

print("Todos los archivos han sido modificados y reemplazados.")
```
